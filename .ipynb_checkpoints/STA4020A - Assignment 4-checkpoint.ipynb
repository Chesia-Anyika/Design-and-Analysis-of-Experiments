## 1.2 Number of Treatment combinations

In a factorial experiment, the number of treatment combinations is the product of the levels of all factors. Here we have:

-   Factor A: 2 levels (Low, High)

-   Factor B: 2 levels (Low, High)

-   Factor C: 2 levels (Low, High)

Thus the number of treatment combinations is:

$$
2 \times 2 \times 2 = 8
$$

## 1.3 Full-Factorial Design Matrix

The full factorial design matrix includes **all possible combinations of the levels of the factors.** Here is the design matrix for Factors A, B, and C:

| **Combination** | Factor A (Temperature) | Factor B (Humidity) | Factor C (Light Exposure) |
|-----------------|------------------|-----------------|---------------------|
| **1**           | Low                    | Low                 | Low                       |
| **2**           | High                   | Low                 | Low                       |
| **3**           | Low                    | High                | Low                       |
| **4**           | High                   | High                | Low                       |
| **5**           | Low                    | Low                 | High                      |
| **6**           | High                   | Low                 | High                      |
| **7**           | Low                    | High                | High                      |
| **8**           | High                   | High                | High                      |

I defined the function `create_full_factorial_design` to create a 2-level full factorial design plan, as per the above table.

```python
def create_full_factorial_design(factors, randomize=False):
   
    # Create a 2-level full factorial design
    design = fullfact([2]*len(factors))
    
    # Convert levels from 0/1 to -1/+1
    design = 2*design - 1
    
    # Convert design matrix to a DataFrame
    df = pd.DataFrame(design, columns=factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
   
    return df

factors = ['A', 'B', 'C']
df = create_full_factorial_design(factors, randomize = False)

# Save the design to an Excel file
df.to_excel('full_factorial_design.xlsx', index=False)

#View the resulting experimental design
print(df)
```

The specific levels within the factors A, B and C are coded as -1 for Low and +1 for High, which is common practice in factorial designs to facilitate the analysis.  

## 1.4 Simulation and Analysis of Response Data
### 1.4.1 Simulation of Response Data
To perform Factorial ANOVA, we need to simulate response data $Y$ for each treatment combination, which I have taken to be **Germination Rate**. First, we define the effect sizes of each factor and all possibles interactions between factors. I defined the effect and interaction sizes such that:

1. **Effect Sizes per Factor:**
   - **Temperature (Factor A):**
     - When the temperature is at the "High" level (1), the response variable is expected to increase by 0.8 units.
     - When the temperature is at the "Low" level (-1), the response variable is expected to increase by 0.5 units.
   - **Humidity (Factor B):**
     - When humidity is at the "High" level (1), the response variable is expected to increase by 0.9 units.
     - When humidity is at the "Low" level (-1), the response variable is expected to increase by 0.6 units.
   - **Light Exposure (Factor C):**
     - When light exposure is at the "High" level (1), the response variable is expected to increase by 0.7 units.
     - When light exposure is at the "Low" level (-1), the response variable is expected to increase by 0.3 units.

2. **Interactions:**
   - **Interaction AB (Temperature and Humidity):**
     - The combined effect of temperature and humidity (when both factors are considered together) results in an additional increase of 0.1 units beyond the individual effects.
   - **Interaction AC (Temperature and Light Exposure):**
     - Considering both temperature and light exposure together, there is an additional increase of 0.2 units beyond their individual effects.
   - **Interaction BC (Humidity and Light Exposure):**
     - When humidity and light exposure are both present, there is an additional increase of 0.15 units beyond their individual effects.
   - **Three-Way Interaction (ABC):**
     - The simultaneous presence of temperature, humidity, and light exposure results in an additional increase of 0.3 units beyond the main effects and pairwise interactions.

```python
# Define effect sizes per factor
effect_A = {-1: 0.5, 1: 0.8}  # Temperature
effect_B = {-1: 0.6, 1: 0.9}  # Humidity
effect_C = {-1: 0.3, 1: 0.7}  # Light Exposure

# Define effect sizes per interaction
interaction_AB = 0.1  # Temperature and Humidity
interaction_AC = 0.2  # Temperature and Light Exposure
interaction_BC = 0.15  # Humidity and Light Exposure
interaction_ABC = 0.3  # Temperature, Humidity, and Light Exposure

# Set seed for reproducibility
np.random.seed(348)

# Number of samples
num_samples = 100

# Create factorial design
design = pd.DataFrame({
    'A': np.tile([-1, 1], num_samples // 2),
    'B': np.tile(np.repeat([-1, 1], num_samples // 4), 2),
    'C': np.repeat([-1, 1], num_samples // 2)
})

# Simulate the response data (germination rate Y)
design['Y'] = (design['A'].map(effect_A)
               + design['B'].map(effect_B)
               + design['C'].map(effect_C)
               + (interaction_AB * ((design['A'] == 1) & (design['B'] == 1)))
               + (interaction_AC * ((design['A'] == 1) & (design['C'] == 1)))
               + (interaction_BC * ((design['B'] == 1) & (design['C'] == 1)))
               + (interaction_ABC * ((design['A'] == 1) & (design['B'] == 1) & (design['C'] == 1)))
               + np.random.normal(loc=0.5, scale=0.1, size=num_samples)  # Adjust mean and variance as needed
               )

# Save the design to an Excel file
design.to_excel('full_factorial_design_Germination_rate.xlsx', index=False)

# Display the simulated data
print(design)
```

### 1.4.2 Main Effects Plot
I defined a function that plots a **Main Effects Plot**, which is a bar chart displaying the main effects on the response variable Y. Positive values indicate an increase in the response Y when the factor is at its high level, while negative values indicate a decrease.

```python
def main_effects_plot(excel_file, result_column):
    # Load data from excel
    df = pd.read_excel(excel_file)
    
    # Identify factors by excluding the result column
    factors = [col for col in df.columns if col != result_column]
        
    # Calculate main effects
    main_effects = {}
    for factor in factors:
        mean_plus = df[df[factor] == 1][result_column].mean()
        mean_minus = df[df[factor] == -1][result_column].mean()
        main_effects[factor] = mean_plus - mean_minus

    # Plot main effects
    fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figsize as needed
    ax.bar(main_effects.keys(), main_effects.values())

    # Add a zero line (reference line)
    ax.axhline(y=0, color='black', linestyle='--', linewidth=1)

    # Annotate the bars with their values
    for factor, value in main_effects.items():
        ax.text(factor, value + 0.01 * abs(value), '{:.2f}'.format(value), ha='center', va='bottom')

    ax.set_ylabel('Main Effect')
    ax.set_title('Main Effects Plot')
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()  # Ensures labels are not cut off
    plt.show()

# Execute function
excel_file = 'full_factorial_design_Germination_rate.xlsx'
result_column = 'Y'
main_effects_plot(excel_file, result_column)
```

**Interpretation**

The plot shows the response variable is affected positively by all the factors, indicating an increase in germination rate when al factors are at their high level. 

The factor Light Exposure (C) has the greatest positive effect size on the response variable, with an effect size of $0.66$. 

### 1.4.3 Interactions Plot
I defined another function to explore two-way interactions in our factorial design. 

```python
#define interactions plot
def interaction_point_plot(excel_file, result_column):
    df = pd.read_excel(excel_file)
    factors = [col for col in df.columns if col != result_column]
    interactions = list(itertools.combinations(factors, 2))

    # Calculate number of rows and columns for subplots
    cols = 3
    rows = math.ceil(len(interactions) / cols)
    fig, axs = plt.subplots(rows, cols, figsize=(10, 10*rows/3))

    for idx, interaction in enumerate(interactions):
        row = idx // cols
        col = idx % cols

        ax = axs[row, col] if rows > 1 else axs[col]

        for level in [-1, 1]:
            subset = df[df[interaction[0]] == level]
            ax.plot(subset[interaction[1]].unique(), subset.groupby(interaction[1])[result_column].mean(), 'o-', label=f'{interaction[0]} = {level}')

        ax.set_title(f'{interaction[0]} x {interaction[1]}')
        ax.legend()
        ax.grid(True)

    # Handle any remaining axes (if there's no data to plot in them)
    for idx in range(len(interactions), rows*cols):
        row = idx // cols
        col = idx % cols
        axs[row, col].axis('off')

    plt.tight_layout()
    plt.show()
    
#Execute function
interaction_point_plot(excel_file, result_column)
```

**Interpretation**
1. **Interaction Plot for Factors A and B:**
   - This plot shows the interaction between temperature (Factor A) and humidity (Factor B).
   - The blue line represents Factor A at the "Low" level (A = -1), while the orange line represents Factor A at the "High" level (A = 1).
   - As humidity (Factor B) increases along the x-axis:
     - For A = -1 (blue line), the response variable (y-axis) increases gradually.
     - For A = 1 (orange line), the response variable increases more steeply.
   - Interpretation: The effect of humidity on the response depends on the level of temperature. When temperature is high (A = 1), humidity has a stronger impact on the response.

2. **Interaction Plot for Factors A and C:**
   - This plot shows the interaction between temperature (Factor A) and light exposure (Factor C).
   - Similar to the previous plot, blue line represents A = -1, and orange line represent A = 1.
   - As light exposure (Factor C) increases along the x-axis:
     - Both levels of Factor A show consistent increases in the response variable.
   - Interpretation: Regardless of temperature level, changes in light exposure have a similar effect on the response.

3. **Interaction Plot for Factors B and C:**
   - This plot represents the interaction between humidity (Factor B) and light exposure (Factor C).
   - The blue line corresponds to B = -1, and the orange line corresponds to B = 1.
   - As light exposure (Factor C) increases:
     - For B = -1 (blue line), the response variable increases gradually.
     - For B = 1 (orange line), the response variable increases more steeply.
   - Interpretation: The impact of light exposure on the response depends on the level of humidity. When humidity is high (B = 1), light exposure has a stronger effect.

## 1.5 Full-Factorial ANOVA Model
After getting an overview of the data, we can conduct the ANOVA model building process. The process involves fitting a linear model to the data and then conducting ANOVA to test the significance of each factor and interaction. We do this to test the assumptions we built from the visualization step.

```python
df = pd.read_excel(excel_file)

# Fit the model
formula = 'Y ~ A*B*C'
model = ols(formula, data=df).fit()

# Perform ANOVA and print the results
anova_table = sm.stats.anova_lm(model, typ=1)
print(anova_table)
```

**Interpretation**
Let us examine the P-Values for each factor and their interactions:

2. **B ($6.025245e^{-56}$):** The p-value is less than the conventional significance level of $0.05$, indicating that factor B also has a significant effect on Y.

3. **A:B ($1.886524e^{-12}$):** The p-value is less than the conventional significance level of $0.05$, indicating a significant interaction effect between factors A and B on Y.

4. **C ($9.220673e^{-55}$):** The p-value is less than the conventional significance level of $0.05$, indicating that factor C has a significant effect on Y.

5. **A:C ($1.495843e^{-19}$):** The p-value is less than the conventional significance level of $0.05$, indicating a significant interaction effect between factors A and C on Y.

6. **B:C ($8.422367e^{-10}$):** The p-value is less than the conventional significance level of $0.05$, indicating a significant interaction effect between factors B and C on Y.

7. **A:B:C ($1.153675e^{-13}$):** The p-value is less than the conventional significance level of $0.05$, indicating a significant three-way interaction effect between factors A, B, and C on Y.

Thus, all the factors, factor pairs and the factor triple significantly affect the outcome response Y. 

This indicates that Temperature, Humidity and Light Exposure as well as all their interactions significantly affect the outcome of response Y. This also validates the effects and interactions we observed using the **Main Effects Plot** and **Interaction Plots**. 

## 1.6 Model Diagnostics
I defined the function `diagnostic_plots` that creates three diagnostic plots to assess the fit of our model. The model creates three plots, which are: 

**Residuals vs. Predicted plot:**

This is a scatter plot of predicted values against residuals. Ideally, the points should be randomly dispersed around the horizontal line y=0, without any discernible pattern.

**Residuals vs. Runs plot**

This plot checks for any patterns in the residuals that may suggest dependence on the order of data collection, which could indicate time-related trends or serial correlation.

**Q-Q plot**

This is used to assess the normality of the residuals. Points following a straight line (the 45-degree line indicated by line='45') suggest that the residuals are normally distributed.

```python
def diagnostic_plots(model):
    # Extract residuals and predicted values from the model
    residuals = model.resid
    predicted = model.fittedvalues
    
    # Create subplots
    fig, axs = plt.subplots(1, 3, figsize=(10, 5))
    
    # Residuals vs Predicted
    axs[0].scatter(predicted, residuals, edgecolors='k', facecolors='none')
    axs[0].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[0].set_title('Residuals vs. Predicted')
    axs[0].set_xlabel('Predicted values')
    axs[0].set_ylabel('Residuals')
    
    # Residuals vs. Runs (Order of Data Collection)
    axs[1].scatter(range(len(residuals)), residuals, edgecolors='k', facecolors='none')
    axs[1].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[1].set_title('Residuals vs. Run')
    axs[1].set_xlabel('Run')
    axs[1].set_ylabel('Residuals')
    
    # Q-Q plot
    sm.qqplot(residuals, line='45', fit=True, ax=axs[2])
    axs[2].set_title('Q-Q Plot')
    
    plt.tight_layout()
    plt.show()
    
#execute function
diagnostic_plots(model)
```

**Interpretation**
1. **Residuals vs. Predicted Plot**
    - The points appear to be fairly randomly dispersed around the y=0y=0 line, suggesting that the residuals do not exhibit any obvious patterns. This is a good indication that the model does not suffer from heteroscedasticity or non-linearity in the residuals.
    
2. **Residuals vs. Run Plot**
    - The residuals appear to be randomly scattered, without any obvious pattern or trend over the order of data collection. This suggests that there is no significant time-related trend or serial correlation affecting the residuals.
    
3. **Q-Q Plot**
    - Most of the points fall along the 45-degree line, suggesting that the residuals are approximately normally distributed. There are some deviations at the tails, which might indicate slight departures from normality, but overall the normality assumption appears reasonably met.
    
Overall, the plots suggest a food fit of the model.

## 2.2 Split-Plot Structure

In a split-plot design, the main plots (whole plots) receive the main plot factor treatments, and each main plot is divided into subplots, which receive the subplot factor treatments.

Main Plot Factor: Fertilizer Type (2 levels: Organic and Inorganic)
- Organic
- Inorganic

Subplot Factor: Irrigation Frequency (3 levels: Low, Medium, High)
- Low
- Medium
- High

## 2.3 Split-Plot Design Matrix

The split-plot design matrix includes all possible combinations of the levels of the factors, structured to show main plots and subplots. I recoded the Fertilizer and Irrigation levels to numbers, to enable analysis using an ANOVA model. The split plot design is as below: 

| **Main Plot** | **Subplot** | **Fertilizer Type (F)** | **Irrigation Frequency (I)** |
|---------------|-------------|---------------------|--------------------------|
| **1**         | **-1**       | Organic             | Low                      |
| **1**         | **0**       | Organic             | Medium                   |
| **1**         | **1**       | Organic             | High                     |
| **-1**         | **-1**       | Inorganic           | Low                      |
| **-1**         | **0**       | Inorganic           | Medium                   |
| **-1**         | **1**       | Inorganic           | High                     |

I defined the function `create_split_plot_design` to create the split plot design as per the above table.

```python
def create_split_plot_design(main_plot_factors, subplot_factors, num_replicates=1, randomize=False):
    # Generate main plot levels
    main_plot_levels = [1, -1]  # 1 for Organic, -1 for Inorganic
    
    # Generate subplot levels
    subplot_levels = [-1, 0, 1]  # -1 for Low, 0 for Medium, 1 for High
    
    # Create a list to store the design
    design = []
    
    for main_plot in main_plot_levels:
        for _ in range(num_replicates):
            for subplot in subplot_levels:
                design.append([main_plot, subplot])
    
    # Convert to DataFrame
    df = pd.DataFrame(design, columns=main_plot_factors + subplot_factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
    
    return df

main_plot_factors = ['F']
subplot_factors = ['I']
num_replicates = 1  # Adjust the number of replicates to get the desired number of samples

df = create_split_plot_design(main_plot_factors, subplot_factors, num_replicates, randomize=True)

# Save the design to an Excel file
df.to_excel('split_plot_design.xlsx', index=False)

# View the resulting experimental design
print(df)
```

## 2.4 Simulation and Analysis of Response Data

### 2.4.1 Simulation of Response Data

To perform the split-plot ANOVA, we need to simulate response data $Y$, which I took to be **Yield**, for each treatment combination.

I defined my effect and interaction sizes as follows: 

1. **Effect Sizes:**
   - **Effect of Factor F (levels 1 and -1):** Factor F at level 1 increases the outcome by 1.0 units compared to its baseline, while at level -1, it increases the outcome by 1.2 units.
   - **Effect of Factor I (levels -1, 0, and 1):** Factor I at level -1 decreases the outcome by 0.8 units, at level 0 increases it by 1.0 units (baseline), and at level 1 increases it by 1.3 units.
   
2. **Interaction Size:**
   - **Interaction Effect (F x I):** The interaction between factors F and I increases the outcome by 0.2 units beyond the sum of their individual effects, indicating synergy between these factors in affecting the outcome variable.
   
I used thes effect and interaction sizes to simulate around 200 samples of the Y response. 

```python
# Define effect sizes per factor
effect_F = {'1': 1.0, '-1': 1.2}
effect_I = {'-1': 0.8, '0': 1.0, '1': 1.3}

# Define effect sizes for interaction
interaction_FI = 0.2

# Set seed for reproducibility
np.random.seed(348)

# Number of main plots and subplots
num_main_plots = 2
num_subplots_per_main_plot = 3
num_replicates = 33  # To get around 200 samples in total 

# Create main plot and subplot factors
main_plot_factors = ['F']
subplot_factors = ['I']

# Function to create split plot design
def create_split_plot_design(main_plot_factors, subplot_factors, num_replicates=1, randomize=False):
    # Generate main plot levels
    main_plot_levels = ['1', '-1']
    
    # Generate subplot levels
    subplot_levels = ['-1', '0', '1']
    
    # Create a list to store the design
    design = []
    
    for main_plot in main_plot_levels:
        for _ in range(num_replicates):
            for subplot in subplot_levels:
                design.append([main_plot, subplot])
    
    # Convert to DataFrame
    df = pd.DataFrame(design, columns=main_plot_factors + subplot_factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
    
    return df

# Generate split plot design
df_split_plot = create_split_plot_design(main_plot_factors, subplot_factors, num_replicates, randomize=True)

# Simulate response data
df_split_plot['Y'] = (
    df_split_plot['F'].map(effect_F)
    + df_split_plot['I'].map(effect_I)
    + (interaction_FI * ((df_split_plot['F'] == 'I') & (df_split_plot['I'] == '1')))
    + np.random.normal(loc=0, scale=0.1, size=len(df_split_plot))  # Add some random noise
)

# Save the design to an Excel file
df_split_plot.to_excel('split_plot_design_Yield.xlsx', index=False)

# Display the simulated data
print(df_split_plot)
```

## 2.6 Model Diagnostics

To validate the assumptions of the model, we will create diagnostic plots.

```python
# Define the diagnostic plots function
def diagnostic_plots(model):
    residuals = model.resid
    predicted = model.fittedvalues

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    # Residuals vs. Predicted
    axs[0].scatter(predicted, residuals, edgecolors='k', facecolors='none')
    axs[0].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[0].set_title('Residuals vs. Predicted')
    axs[0].set_xlabel('Predicted values')
    axs[0].set_ylabel('Residuals')

    # Residuals vs. Runs (Order of Data Collection)
    axs[1].scatter(range(len(residuals)), residuals, edgecolors='k', facecolors='none')
    axs[1].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[1].set_title('Residuals vs. Run')
    axs[1].set_xlabel('Run')
    axs[1].set_ylabel('Residuals')

    # Q-Q plot
    sm.qqplot(residuals, line='45', fit=True, ax=axs[2])
    axs[2].set_title('Q-Q Plot')

    plt.tight_layout()
    plt.show()

# Execute the diagnostic plots function
diagnostic_plots(model)
```

1. **Residuals vs. Predicted Plot**
    - The model’s residuals exhibit homoscedasticity, indicating consistent variance across predicted values, despite the discernible linear patterns observed. This suggests a good fit of the model to the data.
    
2. **Residuals vs. Run Plot**
    - The model’s residuals do not show any time-related effects or non-random error terms. This suggests a good fit of the model to the data. 
    
3. **Q-Q Plot**
    -  Residuals approximately follow a normal distribution, but there are minor deviations in the tails. This suggests a good fit of the model.
    
Overall, the plots suggest a good fit of the model to the data.
