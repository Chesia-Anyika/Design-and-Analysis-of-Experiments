## 1.1 Number of Treatment combinations

In a factorial experiment, the number of treatment combinations is the product of the levels of all factors. Here we have:

- Factor A: Diet (3 levels: Low, Medium, High) 

- Factor B: Exercise (2 levels: None, Daily) 

- Factor C: Supplement (2 levels: With, Without) 


Thus the number of treatment combinations is:

$$
3 \times 2 \times 2 = 12
$$

## 1.2 Full-Factorial Design Matrix

The full factorial design matrix includes **all possible combinations of the levels of the factors.** Here is the design matrix for Factors A, B, and C:

| **Combination** | Factor A (Diet) | Factor B (Exercise) | Factor C (Supplement) |
|-----------------|------------------|---------------------|-----------------------|
| **1**           | Low              | None                | With                  |
| **2**           | Low           | None                | Without                  |
| **3**           | Low             | Daily                | With                  |
| **4**           | Low              | Daily               | Without                  |
| **5**           | Medium           | None               | With                  |
| **6**           | Medium             | None               | Without                  |
| **7**           | Medium              | Daily                | With               |
| **8**           | Medium           | Daily                | Without               |
| **9**           | High             | None                | With               |
| **10**          | High              | None               | Without               |
| **11**          | High           | Daily               | With               |
| **12**          | High             | Daily               | Without               |

I defined the function `create_full_factorial_design` to create a full factorial design plan, as per the above table. I recoded the factors to numerical entries to facilitate modelling of the data, as follows: 

**Factor A (Diet)**
- Low (-1)
- Medium (0)
- High (1)

**Factor B (Exercise)**
- None (-1)
- Daily (1)

**Factor C (Supplement)**
- With (-1)
- Without (1)

```python
def create_full_factorial_design(factors, levels, randomize=False):
    # Generate the full factorial design using itertools.product
    design = list(itertools.product(*levels))
    
    # Convert design matrix to a DataFrame
    df = pd.DataFrame(design, columns=factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
    
    return df

# Define the factors and their levels
factors = ['A', 'B', 'C']
levels = [
    [-1, 0, 1],  # Diet: Low (-1), Medium (0), High (1)
    [-1, 1],     # Exercise: None (-1), Daily (1)
    [-1, 1]      # Supplement: With (-1), Without (1)
]

# Create the factorial design
df = create_full_factorial_design(factors, levels, randomize=False)

# Save the design to an Excel file
df.to_excel('full_factorial.xlsx', index=False)

# View the resulting experimental design
print(df)
```

## 1.3 Simulation and Analysis of Response Data
### 1.3.1 Simulation of Response Data
To perform Factorial ANOVA, we need to simulate response data $Y$ for each treatment combination, which is **Health Outcome**. First, we define the effect sizes of each factor and all possibles interactions between factors. I defined the effect and interaction sizes such that:

1. **Effect Sizes per Factor:**
   - **Diet (Factor A):**
     - When Diet is at the "High" level (1), the response variable is expected to increase by 0.8 units.
     - When Diet is at the "Medium" level (1), the response variable is expected to increase by 0.8 units.     
     - When Diet is at the "Low" level (-1), the response variable is expected to increase by 0.5 units.
   - **Exercise (Factor B):**
     - When exercise is at the "Daily" level (1), the response variable is expected to increase by 0.9 units.
     - When exercise is at the "None" level (-1), the response variable is expected to increase by 0.6 units.
   - **Supplement (Factor C):**
     - When supplement is at the "With" level (1), the response variable is expected to increase by 0.7 units.
     - When supplement is at the "Without" level (-1), the response variable is expected to increase by 0.3 units.

2. **Interactions:**
   - **Interaction AB (Diet and Exercise):**
     - The combined effect of diet and exercise (when both factors are considered together) results in an additional increase of 0.1 units beyond the individual effects.
   - **Interaction AC (Exercise and Supplement):**
     - Considering both Exercise and Supplement together, there is an additional increase of 0.2 units beyond their individual effects.
   - **Interaction BC (Humidity and Light Exposure):**
     - When exercise and supplement are both present, there is an additional increase of 0.15 units beyond their individual effects.
   - **Three-Way Interaction (ABC):**
     - The simultaneous presence of diet, exercise and supplement results in an additional increase of 0.3 units beyond the main effects and pairwise interactions.

```python
# Define the function to create the full factorial design
def create_full_factorial_design(factors, levels, randomize=False):
    # Generate the full factorial design using itertools.product
    design = list(itertools.product(*levels))
    
    # Convert design matrix to a DataFrame
    df = pd.DataFrame(design, columns=factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
    
    return df

# Define the factors and their levels
factors = ['A', 'B', 'C']  # A = Diet, B = Exercise, C = Supplement
levels = [
    [-1, 0, 1],  # A: Low (-1), Medium (0), High (1)
    [-1, 1],     # B: None (-1), Daily (1)
    [-1, 1]      # C: With (-1), Without (1)
]

# Create the factorial design
base_df = create_full_factorial_design(factors, levels, randomize=False)

# Number of samples needed
num_samples = 100

# Define effect sizes per factor (realistic values)
effect_A = {-1: 0.2, 0: 0.5, 1: 0.8}  # A: Diet (Low, Medium, High)
effect_B = {-1: 0.3, 1: 0.7}          # B: Exercise (None, Daily)
effect_C = {-1: 0.4, 1: 0.6}          # C: Supplement (With, Without)

# Define effect sizes per interaction (realistic values)
interaction_AB = 0.05  # A and B
interaction_AC = 0.07  # A and C
interaction_BC = 0.06  # B and C
interaction_ABC = 0.1  # A, B, and C

# Set seed for reproducibility
np.random.seed(348)

# Repeat the base design enough times to get at least num_samples
df = pd.concat([base_df] * (num_samples // len(base_df) + 1), ignore_index=True)
df = df.sample(n=num_samples, random_state=348).reset_index(drop=True)

# Simulate the response data (Health Outcome)
df['Y'] = (
    df['A'].map(effect_A)
    + df['B'].map(effect_B)
    + df['C'].map(effect_C)
    + (interaction_AB * ((df['A'] != 0) & (df['B'] == 1)))
    + (interaction_AC * ((df['A'] != 0) & (df['C'] == 1)))
    + (interaction_BC * ((df['B'] == 1) & (df['C'] == 1)))
    + (interaction_ABC * ((df['A'] != 0) & (df['B'] == 1) & (df['C'] == 1)))
    + np.random.normal(loc=0, scale=0.1, size=num_samples)  # Adjust mean and variance as needed
)

# Save the design to an Excel file
df.to_excel('full_factorial_health_outcome.xlsx', index=False)

# Display the simulated data
print(df)
```

### 1.3.2 Main Effects Plot
I defined a function that plots a **Main Effects Plot**, which is a bar chart displaying the main effects on the response variable Y. Positive values indicate an increase in the response Y when the factor is at its high level, while negative values indicate a decrease.

```python
# Define the function to create the main effects plot
def main_effects_plot(excel_file, result_column):
    # Load data from excel
    df = pd.read_excel(excel_file)
    
    # Identify factors by excluding the result column
    factors = [col for col in df.columns if col != result_column]
        
    # Calculate main effects
    main_effects = {}
    for factor in factors:
        if len(df[factor].unique()) == 2:
            mean_plus = df[df[factor] == 1][result_column].mean()
            mean_minus = df[df[factor] == -1][result_column].mean()
            main_effects[factor] = mean_plus - mean_minus
        elif len(df[factor].unique()) == 3:
            mean_low = df[df[factor] == -1][result_column].mean()
            mean_medium = df[df[factor] == 0][result_column].mean()
            mean_high = df[df[factor] == 1][result_column].mean()
            main_effects[factor] = {
                'Low-Medium': mean_low - mean_medium,
                'Medium-High': mean_medium - mean_high,
                'Low-High': mean_low - mean_high
            }

    # Plot main effects
    fig, ax = plt.subplots(figsize=(12, 8))  # Adjust figsize as needed

    # Plot factors with two levels
    two_level_effects = {k: v for k, v in main_effects.items() if not isinstance(v, dict)}
    if two_level_effects:
        ax.bar(two_level_effects.keys(), two_level_effects.values())

    # Plot factors with three levels
    three_level_effects = {k: v for k, v in main_effects.items() if isinstance(v, dict)}
    for i, (factor, effects) in enumerate(three_level_effects.items()):
        for key, val in effects.items():
            ax.bar(f"{factor} {key}", val)

    # Add a zero line (reference line)
    ax.axhline(y=0, color='black', linestyle='--', linewidth=1)

    # Annotate the bars with their values
    for factor, value in main_effects.items():
        if isinstance(value, dict):
            for key, val in value.items():
                ax.text(f"{factor} {key}", val + 0.01 * abs(val), '{:.2f}'.format(val), ha='center', va='bottom')
        else:
            ax.text(factor, value + 0.01 * abs(value), '{:.2f}'.format(value), ha='center', va='bottom')

    ax.set_ylabel('Main Effect')
    ax.set_title('Main Effects Plot')
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()  # Ensures labels are not cut off
    plt.show()

# Execute function
excel_file = 'full_factorial_health_outcome.xlsx'
result_column = 'Y'
main_effects_plot(excel_file, result_column)
```

**Interpretation**
1. **Factor B - Exercise**: This bar has a positive main effect value of 0.44. As Daily exercise is coded as 1, this indicates that Daily Exercise has a strong positive impact on the health outcomes, rather than None. 

2. **Factor C - Supplement**: This bar has a positive main effect value of 0.25. As 'Without' Supplements is coded as 1, this shows that Not taking supplements has a positive impact on the health outcomes, to a lesser extent than Exercise.

3. **Low and Medium Diets**: The orange bar represents the interaction between factors Low and Medium Diets. It has a negative main effect value of -0.19, suggesting that the Low intensity diets have a higher negative impact on health outcomes than medium intensity diets.

4. **Medium and High Diets**: The green bar represents the interaction between factors Low and Medium Diets. It has a negative main effect value of -0.37, suggesting that the Medium intensity diets have a higher negative impact on health outcomes than High intensity diets, to a higher extent than the low-Medium pair.

5. **Low and High Diets**: The red bar represents the interaction between factors Low and High Diets. It has a negative main effect value of -0.56, suggesting that the Low intensity diets have a higher negative impact on health outcomes than High intensity diets, to higher extent than the other diet pairs.

### 1.3.3 Interactions Plot
I defined another function to explore two-way interactions in our factorial design. 

```python
# Define the function to create the interaction plot
def interaction_point_plot(excel_file, result_column):
    df = pd.read_excel(excel_file)
    factors = [col for col in df.columns if col != result_column]
    interactions = list(itertools.combinations(factors, 2))

    # Calculate number of rows and columns for subplots
    cols = 3
    rows = math.ceil(len(interactions) / cols)
    fig, axs = plt.subplots(rows, cols, figsize=(10, 10 * rows / 3))

    for idx, interaction in enumerate(interactions):
        row = idx // cols
        col = idx % cols

        ax = axs[row, col] if rows > 1 else axs[col]

        # Plot for each combination of levels
        for level in sorted(df[interaction[0]].unique()):
            subset = df[df[interaction[0]] == level]
            means = subset.groupby(interaction[1])[result_column].mean()
            ax.plot(means.index, means, 'o-', label=f'{interaction[0]} = {level}')

        ax.set_title(f'{interaction[0]} x {interaction[1]}')
        ax.legend()
        ax.grid(True)

    # Handle any remaining axes (if there's no data to plot in them)
    for idx in range(len(interactions), rows * cols):
        row = idx // cols
        col = idx % cols
        axs[row, col].axis('off')

    plt.tight_layout()
    plt.show()

# Execute the interaction plot function
interaction_point_plot(excel_file, result_column)
```

**Interpretation**
1. **A x B Interaction Plot**: This plot shows the interaction between variables A and B. The lines for A = -1 and A = 0 are parallel, suggesting no interaction effect between A and B at these levels. However, the line for A = 1 has a different slope, indicating an interaction effect where the relationship between B and the dependent variable changes when A is at level 1.

2. **A x C Interaction Plot**: This plot shows the interaction between variables A and C. All three lines have different slopes indicating that there is an interaction effect between variables A and C at all levels of A. This means the effect of A on the dependent variable depends on the level of C and vice versa.

3. **B x C Interaction Plot**: This plot shows the interaction between variables B and C. Both lines have different slopes which suggest that there is an interaction effect between variables B and C at both levels of B. This means the effect of B on the dependent variable depends on the level of C and vice versa.

## 1.5 Model Diagnostics
I defined the function `diagnostic_plots` that creates three diagnostic plots to assess the fit of our model. The model creates three plots, which are: 

**Residuals vs. Predicted plot:**

This is a scatter plot of predicted values against residuals. Ideally, the points should be randomly dispersed around the horizontal line y=0, without any discernible pattern.

**Residuals vs. Runs plot**

This plot checks for any patterns in the residuals that may suggest dependence on the order of data collection, which could indicate time-related trends or serial correlation.

**Q-Q plot**

This is used to assess the normality of the residuals. Points following a straight line (the 45-degree line indicated by line='45') suggest that the residuals are normally distributed.

```python
def diagnostic_plots(model):
    # Extract residuals and predicted values from the model
    residuals = model.resid
    predicted = model.fittedvalues
    
    # Create subplots
    fig, axs = plt.subplots(1, 3, figsize=(10, 5))
    
    # Residuals vs Predicted
    axs[0].scatter(predicted, residuals, edgecolors='k', facecolors='none')
    axs[0].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[0].set_title('Residuals vs. Predicted')
    axs[0].set_xlabel('Predicted values')
    axs[0].set_ylabel('Residuals')
    
    # Residuals vs. Runs (Order of Data Collection)
    axs[1].scatter(range(len(residuals)), residuals, edgecolors='k', facecolors='none')
    axs[1].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[1].set_title('Residuals vs. Run')
    axs[1].set_xlabel('Run')
    axs[1].set_ylabel('Residuals')
    
    # Q-Q plot
    sm.qqplot(residuals, line='45', fit=True, ax=axs[2])
    axs[2].set_title('Q-Q Plot')
    
    plt.tight_layout()
    plt.show()
    
#execute function
diagnostic_plots(model)
```

**Interpretation**
1. **Residuals vs. Predicted Plot**
    - The points appear to be fairly randomly dispersed around the y=0y=0 line, suggesting that the residuals do not exhibit any obvious patterns. This is a good indication that the model does not suffer from heteroscedasticity or non-linearity in the residuals.
    
2. **Residuals vs. Run Plot**
    - The residuals appear to be randomly scattered, without any obvious pattern or trend over the order of data collection. This suggests that there is no significant time-related trend or serial correlation affecting the residuals.
    
3. **Q-Q Plot**
    - Most of the points fall along the 45-degree line, suggesting that the residuals are approximately normally distributed. There are some deviations at the tails, which might indicate slight departures from normality, but overall the normality assumption appears reasonably met.
    
Overall, the plots suggest a food fit of the model.

## 2.1 Split-Plot Structure

In a split-plot design, the main plots (whole plots) receive the main plot factor treatments, and each main plot is divided into subplots, which receive the subplot factor treatments.

Main Plot Factor: Soil Type (2 levels: Sandy and Clay)
- Sandy
- Clay

Subplot Factor: Watering Frequency (3 levels: Low, Medium, High)
- Low
- Medium
- High

## 2.2 Split-Plot Design Matrix

The split-plot design matrix includes all possible combinations of the levels of the factors, structured to show main plots and subplots. I recoded the Soil Type and Watering Frequency levels to numbers, to enable analysis using an ANOVA model. The split plot design is as below: 

| **Main Plot** | **Subplot** | **Soil Type (S)** | **Watering Frequency (W)** |
|---------------|-------------|---------------------|--------------------------|
| **1**         | **-1**       | Sandy             | Low                      |
| **1**         | **0**       | Sandy             | Medium                   |
| **1**         | **1**       | Sandy             | High                     |
| **-1**         | **-1**       | Clay           | Low                      |
| **-1**         | **0**       | Clay           | Medium                   |
| **-1**         | **1**       | Clay           | High                     |

I defined the function `create_split_plot_design` to create the split plot design as per the above table.

```python
def create_split_plot_design(main_plot_factors, subplot_factors, num_replicates=1, randomize=False):
    # Generate main plot levels
    main_plot_levels = [1, -1]  # 1 for Sandy, -1 for Clay
    
    # Generate subplot levels
    subplot_levels = [-1, 0, 1]  # -1 for Low, 0 for Medium, 1 for High
    
    # Create a list to store the design
    design = []
    
    for main_plot in main_plot_levels:
        for _ in range(num_replicates):
            for subplot in subplot_levels:
                design.append([main_plot, subplot])
    
    # Convert to DataFrame
    df = pd.DataFrame(design, columns=main_plot_factors + subplot_factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
    
    return df

main_plot_factors = ['S']
subplot_factors = ['W']
num_replicates = 1  # Adjust the number of replicates to get the desired number of samples

df = create_split_plot_design(main_plot_factors, subplot_factors, num_replicates, randomize=True)

# Save the design to an Excel file
df.to_excel('split_plot_design.xlsx', index=False)

# View the resulting experimental design
print(df)
```

## 2.3 Simulation and Analysis of Response Data

### 2.3.1 Simulation of Response Data

To perform the split-plot ANOVA, we need to simulate response data $Y$, which is**Plant Growth**, for each treatment combination.

I defined my effect and interaction sizes as follows: 

1. **Effect Sizes:**
   - **Effect of Factor S (levels 1 and -1):** Factor F at level 1 increases the outcome by 1.0 units compared to its baseline, while at level -1, it increases the outcome by 1.2 units.
   - **Effect of Factor W (levels -1, 0, and 1):** Factor I at level -1 decreases the outcome by 0.8 units, at level 0 increases it by 1.0 units (baseline), and at level 1 increases it by 1.3 units.
   
2. **Interaction Size:**
   - **Interaction Effect (S x W):** The interaction between factors F and I increases the outcome by 0.2 units beyond the sum of their individual effects, indicating synergy between these factors in affecting the outcome variable.
   
I used thes effect and interaction sizes to simulate around 200 samples of the Y response. 

```python
import numpy as np
import pandas as pd

# Define effect sizes per factor
effect_S = {-1: 0.8, 1: 1.2}  # Clay (-1) vs. Sandy (1)
effect_W = {-1: 0.7, 0: 1.0, 1: 1.5}  # Low (-1), Medium (0), High (1)

# Define effect sizes for interaction
interaction_SW = 0.3  # Effect of soil type and watering frequency interaction

# Set seed for reproducibility
np.random.seed(348)

# Number of main plots and subplots
num_replicates = 33  # To get around 200 samples in total 

# Create main plot and subplot factors
main_plot_factors = ['S']
subplot_factors = ['W']

# Function to create split plot design
def create_split_plot_design(main_plot_factors, subplot_factors, num_replicates=1, randomize=False):
    # Generate main plot levels
    main_plot_levels = [-1, 1]
    
    # Generate subplot levels
    subplot_levels = [-1, 0, 1]
    
    # Create a list to store the design
    design = []
    
    for main_plot in main_plot_levels:
        for _ in range(num_replicates):
            for subplot in subplot_levels:
                design.append([main_plot, subplot])
    
    # Convert to DataFrame
    df = pd.DataFrame(design, columns=main_plot_factors + subplot_factors)
    
    # Randomize the design if needed
    if randomize:
        df = df.sample(frac=1).reset_index(drop=True)
    
    return df

# Generate split plot design
df_split_plot = create_split_plot_design(main_plot_factors, subplot_factors, num_replicates, randomize=True)

# Simulate response data
df_split_plot['Y'] = (
    df_split_plot['S'].map(effect_S)
    + df_split_plot['W'].map(effect_W)
    + (interaction_SW * ((df_split_plot['S'] == 1) & (df_split_plot['W'] == 1)))
    + np.random.normal(loc=0, scale=0.1, size=len(df_split_plot))  # Add some random noise
)

# Save the design to an Excel file
df_split_plot.to_excel('split_plot_design_Growth.xlsx', index=False)

# Display the simulated data
print(df_split_plot)
```

### 2.3.3 Interactions Plot
I defined another function to explore two-way interactions in our split plot design. 

```python
def interaction_point_plot(excel_file, result_column):
    df = pd.read_excel(excel_file)
    factors = [col for col in df.columns if col != result_column]
    interactions = list(itertools.combinations(factors, 2))

    if len(interactions) != 1:
        raise ValueError("This function is intended for exactly one interaction plot.")

    interaction = interactions[0]  # Only one interaction to plot

    fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figsize as needed

    for level in [-1, 1]:
        subset = df[df[interaction[0]] == level]
        ax.plot(subset[interaction[1]].unique(), subset.groupby(interaction[1])[result_column].mean(), 'o-', label=f'{interaction[0]} = {level}')

    ax.set_title(f'{interaction[0]} x {interaction[1]}')
    ax.legend()
    ax.grid(True)

    plt.tight_layout()
    plt.show()

# Execute function
excel_file = 'split_plot_design_Growth.xlsx'
result_column = 'Y'
interaction_point_plot(excel_file, result_column)
```

**Interpretations**

The interaction plot between factors S and W in the split-plot design reveals the following patterns. 
- For S = -1, as factor W increases, the response variable also increases.  
- For S = 1, the response initially increases with increasing W.

Thus, different levels of factor S modify how factor W influences the response.

## 2.6 Model Diagnostics

To validate the assumptions of the model, we will create diagnostic plots.

```python
# Define the diagnostic plots function
def diagnostic_plots(model):
    residuals = model.resid
    predicted = model.fittedvalues

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    # Residuals vs. Predicted
    axs[0].scatter(predicted, residuals, edgecolors='k', facecolors='none')
    axs[0].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[0].set_title('Residuals vs. Predicted')
    axs[0].set_xlabel('Predicted values')
    axs[0].set_ylabel('Residuals')

    # Residuals vs. Runs (Order of Data Collection)
    axs[1].scatter(range(len(residuals)), residuals, edgecolors='k', facecolors='none')
    axs[1].axhline(y=0, color='k', linestyle='dashed', linewidth=1)
    axs[1].set_title('Residuals vs. Run')
    axs[1].set_xlabel('Run')
    axs[1].set_ylabel('Residuals')

    # Q-Q plot
    sm.qqplot(residuals, line='45', fit=True, ax=axs[2])
    axs[2].set_title('Q-Q Plot')

    plt.tight_layout()
    plt.show()

# Execute the diagnostic plots function
diagnostic_plots(model)
```

**Interpretation**
1. **Residuals vs. Predicted Plot**
    - The model’s residuals exhibit homoscedasticity, indicating consistent variance across predicted values, despite the discernible linear patterns observed. This suggests a good fit of the model to the data.
    
2. **Residuals vs. Run Plot**
    - The model’s residuals do not show any time-related effects or non-random error terms. This suggests a good fit of the model to the data. 
    
3. **Q-Q Plot**
    -  Residuals approximately follow a normal distribution, but there are minor deviations in the tails. This suggests a good fit of the model.
    
Overall, the plots suggest a good fit of the model to the data.
